```python
import os
import pandas as pd
```

```python


# Specify the source folder and the destination folder
#source_folder = '..\\data\\raw\\data-vn-20230228\\data-vn-20230228\\stock-historical-data'  # Replace with your folder path
#destination_file = '..\\data\\toprocess\\combined_all_close.csv'  # Path for saving the combined file

source_folder = '..\\data\\raw\\data-vn-20230228\\data-vn-20230228\\stock-historical-data'
destination_file = '..\\data\\toprocess\\close_price_vn_data.csv'
# Function to extract the last 7 days of closing prices and return a DataFrame with ticker name as the column
def extract_close_columns(file_path):
    try:
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)

        # Ensure 'TradingDate' column is in dateTradingDate format (adjust column name if necessary)
        df['TradingDate'] = pd.to_datetime(df['TradingDate'], errors='coerce')

        # Sort by TradingDate in ascending order
        df = df.sort_values(by='TradingDate', ascending=True)

        # Drop rows with missing 'Close' or 'TradingDate' values
        df.dropna(subset=['Close', 'TradingDate'], inplace=True)

        # Extract the 'Close' column and keep the 'TradingDate'
        close_col = df[['TradingDate', 'Close']]

        return close_col
    except Exception as e:
        print(f"Error processing file {file_path}: {e}")
        return None

# Initialize an empty DataFrame to combine the Close prices
combined_df = pd.DataFrame()

# Loop through all files in the source folder
for filename in os.listdir(source_folder):
    file_path = os.path.join(source_folder, filename)

    # Check if it's a file (not a directory)
    if os.path.isfile(file_path):
        # Extract the last 7 days of data for the file
        close_data = extract_close_columns(file_path)
        
        if close_data is not None:
            # Get the ticker (file name without the extension)
            ticker = filename.split('-')[0]

            # Rename the 'Close' column to the ticker
            close_data.rename(columns={'Close': ticker}, inplace=True)

            # Merge the data with the combined dataframe
            if combined_df.empty:
                # If the combined dataframe is empty, initialize it with the first file's data
                combined_df = close_data
            else:
                # Otherwise, merge on the 'TradingDate' column
                combined_df = pd.merge(combined_df, close_data, on='TradingDate', how='inner')

# Save the combined DataFrame to a CSV file
combined_df.to_csv(destination_file, index=False)

print(f"Combined data saved to {destination_file}")
```
