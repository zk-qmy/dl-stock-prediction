```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'..\data\toprocess\nasdag_chosen.csv')
df
```

```python
data = df.copy()
data.info()
```

```python
data.describe()
```

```python
data.shape
```

```python
# Count of null values in each column
print("\nCount of nan values in each column:")
print(data.isna().sum())  # or data.isnull().sum()
```

```python
# Visualize the time series
sns.set(style="whitegrid")  # Setting the style to whitegrid for a clean background
 
plt.figure(figsize=(12, 6))  # Setting the figure size
sns.lineplot(data, x='Date', y='High', label='High Price', color='blue')
 
# Adding labels and title
plt.xlabel('Date')
plt.ylabel('High')
plt.title('Share Highest Price Over Time')
 
plt.show()
```
